```python
from cvxopt import matrix, solvers
```

```python
c = matrix([-1., -2.])
G = matrix([[1., -1., -1., 0.], [1., 1., 0., -1.]])
h = matrix([1., 0., 0., 0.])

sol = solvers.lp(c, G, h)
print(sol['x'])
```

```python
c = matrix([-1., -1.])
G = matrix([[6., 1., 2., -1., 0.], [10., -2., -1., 0., -1.]])
h = matrix([47., 2., 7., 0., 0.])

sol = solvers.lp(c, G, h)
print(sol['x'])
```

```python
from cvxopt.glpk import ilp
```

```python
c = matrix([-1., -2.])
G = matrix([[1., -1., -1., 0.], [1., 1., 0., -1.]])
h = matrix([1., 0., 0., 0.])
(status,x) = ilp(c=c,G=G,h=h,I=set(range(2)))

sol = solvers.lp(c, G, h)
print(x)
```
