下記プログラム（問4-2と同じもの）に対して、while 文の実行中に読み込んだ整数が負の値であれば、繰り返しを中断するように if 文と break 文を追加しなさい

```python
score = list()
n = int(input())
count = 0
while count < n:
	s = int(input())
	if s < 0 :
		break
	score.append(s)
	count = count + 1
print(sum(score)/count)
```

下記プログラム（問4-2と同じもの）に対して、while 文の実行中に読み込んだ整数が負の値であれば、その値を無視して処理を継続するよう if 文と continue 文を追加しなさい。

```python
score = list()
n = int(input())
count = 0
while count < n:
	s = int(input())
	if s < 0 :
		continue
	score.append(s)
	count = count + 1
print(sum(score)/count)
```

変数に代入された得点のリストから、while 文を使って各要素を確認し、最大値を見つけて出力するプログラムを作りなさい。

最大値は、以下の手順で求めることができる。

1. リストの先頭の要素を「最大値の候補」として変数に記憶する。
2. リストの各要素を「最大値の候補」と比較して、もしリストの要素のほうが最大値の候補より大きければ、「最大値の候補」である変数の値を更新する。  

max 関数を使わず、自分で対応する処理を作ってみること。

```python
# 以下のリストから最大値を見つけること
scores = [80, 20, 90, 65, 70, 75, 60]
# ここから下にプログラムを記入してください
max = scores[0]
i = 0
while i < len(scores):
	if max < scores[i]:
		max = scores[i]
	i += 1
print(max)
```

整数を 1つ読み取り、その数値が素数であれば True、そうでなければ False を出力するプログラムを作成しなさい。

素数の判定方法：n が 1 より大きい自然数であり、かつ、2からn-1のいずれでも割り切れなければ、n は素数である

```python
x = int(input())
i = 2
cnt = 0
if x <= 1:
	print('False')
elif x == 2:
	print('True')
else:
	while i < x:
		if x % i != 0:
			cnt += 1
		i += 1 
	if x - 2 == cnt:
		print('True')
	else:
		print('False')
```

整数を 1つ読み取り、その数値以下のすべての素数を出力するプログラムを作成しなさい。

数値は1行に1つ出力すればよい。

入出力例：

- 2 が入力されたら： 2 だけを出力する
- 1 が入力されたら：何も出力せずプログラムが終わる
- 10が入力されたら：2, 3, 5, 7 を１行に１つずつ順番に出力する

```python
x = int(input())
if x == 2:
    print(2)
else:
    for i in range(2, x+1):
        if i == 2:
            print(2)
        else:
            cnt = 0
            j = 2
            while j < i:
                if i % j != 0:
                    cnt += 1
                j += 1 
            if i - 2 == cnt:
                print(i)
```

変数に代入された得点リストのそれぞれの要素に対して、

- 得点が90点以上なら "S",
- 90点未満80点以上なら "A",
- 80点未満70点以上なら "B",
- 70点未満60点以上なら "C",
- 60点未満なら "F" を、  

各1行ずつで出力するプログラムを作成しなさい。

```python
scores = [80, 20, 90, 65, 70, 75, 60]
# ここから下にプログラムを記入してください
for i in scores:
	if i >= 90:
		print("S")
	elif 90 > i >= 80:
		print("A")
	elif 80 > i >= 70:
		print("B")
	elif 70 > i >= 60:
		print("C")
	elif 60 > i:
		print("F")
```

変数に代入された得点リストのそれぞれの要素に対して、  

- 得点が90点以上 (成績 "S") の人数,
- 90点未満80点以上(成績 "A")の人数,
- 80点未満70点以上(成績 "B")の人数,
- 70点未満60点以上（成績 "C"）の人数,
- 60点未満(成績 "F")の人数を,  

各1行ずつで、{成績}={人数} の形式で出力するプログラムを作成しなさい。  

初期コードとして与えられているリストの場合、出力は以下のようになる。

S=1  
A=1  
B=2  
C=2  
F=1

```python
# 以下のリストから最大値を見つけること
scores = [80, 20, 90, 65, 70, 75, 60]
# ここから下にプログラムを記入してください
s = 0
a = 0
b = 0
c = 0
f = 0
for i in scores:
	if i >= 90:
		s += 1
	elif 90 > i >= 80:
		a += 1
	elif 80 > i >= 70:
		b += 1
	elif 70 > i >= 60:
		c += 1
	elif 60 > i:
		f += 1

print("S={}".format(s))
print("A={}".format(a))
print("B={}".format(b))
print("C={}".format(c))
print("F={}".format(f))
```

最初に人数 n, そのあと人数分の得点データが順番に入力されるとする。それらを読み込み、

- 得点が90点以上 (成績 "S") の人数,
- 90点未満80点以上(成績 "A")の人数,
- 80点未満70点以上(成績 "B")の人数,
- 70点未満60点以上（成績 "C"）の人数,
- 60点未満(成績 "F")の人数を,  

各1行ずつで、{成績}={人数} の形式で出力するプログラムを作成しなさい。

入力例  

7  
80  
20  
90  
65  
70  
75  
60  

出力例

B=2

```python
n = int(input())
l = []
for i in range(n):
    l.append(int(input()))

s = 0
a = 0
b = 0
c = 0
f = 0
for i in l:
	if i >= 90:
		s += 1
	elif 90 > i >= 80:
		a += 1
	elif 80 > i >= 70:
		b += 1
	elif 70 > i >= 60:
		c += 1
	elif 60 > i:
		f += 1

print("S={}".format(s))
print("A={}".format(a))
print("B={}".format(b))
print("C={}".format(c))
print("F={}".format(f))
```
