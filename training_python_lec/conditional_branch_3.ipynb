入力として、年 y を表現する整数を読み取り、それがうるう年ならば True、そうでなければ False を出力するプログラムを作成しなさい。  

うるう年の判定方法は以下の通りです。  

- y が 400 で割り切れるなら、うるう年である
- y が 400 で割り切れず、100 で割り切れるなら、うるう年ではない
- y が 100 で割り切れないが、4 で割り切れるなら、うるう年である
- y が 4で割り切れないなら、うるう年ではない

```python
 y = int(input())
if y % 400 == 0:
	print('True')
else:
	if y % 100 == 0:
		print('False')
	else:
		if y % 4 == 0:
			print('True')
		else:
			print('False')
```
