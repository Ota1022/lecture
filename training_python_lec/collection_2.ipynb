「問２－３」への解答として作成したプログラムを，ディクショナリを用いたプログラムに書き換えなさい．すなわち，リストscoresをディクショナリscoresに変更しなさい．ただし，ディクショナリscoresにおけるキーは，試験の点数を格納していた４つの変数の名前を用いるものとします．  

なお，ディクショナリに対してはsum関数は使えない．代わりに，「ディクショナリ名.values()」を使ってすべての要素の値（の集まり）をまず取得し，そのうえでsum関数を使う，という方法がある． 

```python
scores = {}
scores['network'] = 88
scores['database'] = 95
scores['security'] = 90
scores['software'] = 100

total = sum(scores.values())
avg = total / len(scores)
print('合計点:{}'.format(total))
print('平均点:{}'.format(avg))
```

キーボードから月日の入力を受け付け，1月1日からその月日までの日数を算出して表示するプログラムを作成しなさい．入力となる「月日」は4桁の数字で与えられ，上位2桁が「月」を下位2桁が「日」を表すものとする．ただし，うるう年ではないとする．  

入出力例  
入力：0115　　　出力：15  
入力：0401　　　出力：91  
入力：1231　　　出力：365  
ヒント  
1月から12月まで，各月の日数を月順に要素として持つリストを作成するとよいかもしれない．

```python
month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
date = input()
ans = 0
if date[:2] == '01':
    ans += int(date[2:4])
else:
    int_month = int(date[:2]) - 1
    int_day = int(date[2:4])
    ans += sum(month[:int_month])
    ans += int_day
print(ans)
```

「課題２－１」で作成したプログラムに，「キーボードから入力した月日の曜日」を合わせて出力する機能を追加しなさい．ただし，1月1日は日曜日とし，うるう年ではないとする．  

入出力例  
入力：0115　 　出力：15 Sunday  
入力：0401　 　出力：91 Saturday  
入力：1231　 　出力：365 Sunday  
ヒント  
日曜日から土曜日まで，各曜日の名称を順に要素として持つリストを作成するとよいかもしれない．

```python
month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
dow = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
date = input()
ans = 0
if date[:2] == '01':
    ans += int(date[2:4])
else:
    int_month = int(date[:2]) - 1
    int_day = int(date[2:4])
    ans += sum(month[:int_month])
    ans += int_day
dow_num = ans % 7 - 1
print(ans, dow[dow_num])
```

京都，大阪，兵庫，奈良の４府県それぞれにおいて，人口の多い３つの市（便宜的に，「三大市」と呼ぶ）とその人口は別紙の通り．このデータを用いて，次のようなプログラムを作成しなさい．  

- 京都，大阪，兵庫の3府県の名称を格納するリストprefecturesを作成する．  

- 3府県それぞれの「三大市の名称をキー，その人口を値とするディクショナリ」を要素とするリストcitiesを作成する．  

- リストprefecturesをキー，リストcitiesを値とするディクショナリpopulationsを，dict(zip())を用いて作成する．  

- キーボードから府県名の入力を受け付け，その府県の三大市の名称と人口合計をディクショナリpopulationsを用いて算出して表示する．  
入出力例  

入力：Kyoto  

出力：Kyoto Uji Kameoka 1719592  

入力：Osaka  

出力：Osaka Sakai Higshiosaka 2058437  

入力：Hyogo  

出力：Kobe Himeji Nishinomiya 2527799  

```python
prefectures = ['Kyoto', 'Osaka', 'Hyogo']
cities = [{'Kyoto' : 1453854, 'Uji' : 179626, 'Kameoka' : 86112} , {'Osaka' : 1463882, 'Sakai' : 362295, 'Higshiosaka' : 232260}, 
{'Kobe': 1513193, 'Himeji' : 527596, 'Nishinomiya': 487010}]
sandai = {'Kyoto' : 'Kyoto Uji Kameoka ', 'Osaka' : 'Osaka Sakai Higshiosaka', 'Hyogo' : 'Kobe Himeji Nishinomiya'}
populations = dict(zip(prefectures, cities))
pre_name = input()
ans_dict = populations[pre_name]
#print(ans_dict.keys(), sum(ans_dict.values()))
print(sandai[pre_name], sum(ans_dict.values()))
```
