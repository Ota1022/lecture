標準ライブラリに含まれるcsvモジュールとsysモジュールが提供する変数や関数などを用いて，「問５－４」と同じ処理を実現するプログラムを書きなさい．

標準入力から与えられる整数値は下図のとおりとする．
- 「試験の点数」を表す値のみで構成され，「試験の点数の個数」を表す値は含まれない．
- 標準入力から与えられる整数値を変数scoresに格納する処理のみを書き換えるものとする．
- 次の代入文によって，標準入力から与えられる値を変数readerに格納することができる．（Readerオブジェクトを生成して変数readerに格納することができる．）  
    　　reader = csv.reader(sys.stdin)  
- Readerオブジェクトの値は，forループで一行ずつ取り出すことができる．例えば次の通り．  
    　　for line in reader:  
    ここで，lineは，標準入力から与えられた一行分の値を格納するリスト．一行が１つの値のみで構成される場合は，第１要素にその値が格納されている．

```python
import csv, sys
reader = csv.reader(sys.stdin)

def get_scores():
	scorelists = list()
	scores = list()
	count = 0
	for line in reader:
		scorelists.append(line)
	for l in scorelists:
		for i in l:
			scores.append(int(i))
	return scores

scores = get_scores()
global threshold
threshold = 60

def calc_total(scores):
	total = 0
	for i in scores: 
		if i >= threshold:
			total += i
	return total

def calc_count(scores):
	count = 0
	for i in scores: 
		if i >= threshold:
			count += 1
	return count

total = calc_total(scores)
count = calc_count(scores)

print("合計点:{}".format(total))
print("平均点:{}".format(total/count))
```

「問７－１」で作成したプログラムを次のように書き換えなさい．

- 標準入力から与えられる整数値は下図のとおりとする．
    - 学生１名分の「試験の点数」が，１行に「カンマ」で区切られて列挙されている．「試験の点数の個数」を表す値は含まれない．
- 標準入力から与えられる整数値を変数scoresに格納する処理のみを書き換えるものとする．
- 「問７－１」での説明と同様，Readerオブジェクトの値は，forループで一行ずつ取り出すことができる．  
    　　for line in reader:  
ここで，lineは，標準入力から与えられた一行分の値を格納するリスト．（標準の設定では）「カンマ」で区切られた値が，行の先頭から順にリストに格納されている．（「カンマ」は取り除かれている．）

```python
import csv, sys
reader = csv.reader(sys.stdin)

def get_scores():
	scores = []
	for line in reader:
		int_line = []
		for i in line:
			int_line.append(int(i))
		scores.append(int_line)
	return scores

scores = get_scores()
n = len(scores)

global threshold
threshold = 0

def calc_total(scores):
	total = 0
	for i in scores: 
		if i >= threshold:
			total += i
	return total

def calc_count(scores):
	count = 0
	for i in scores: 
		if i >= threshold:
			count += 1
	return count


for i in range(n):
	total = calc_total(scores[i])
	count = calc_count(scores[i])

	print("合計点:{}".format(total))
	print("平均点:{}".format(total/count))
```

標準ライブラリに含まれるrandomモジュール，および，外部ライブラリmatplotlibに含まれるpyplotモジュールが提供する変数や関数などを用いて，次のような処理を行うプログラムを書きなさい．

- 0以上100以下の整数値をランダムに200個生成し，「試験の点数」としてリストscoresに格納する．
- リストscoresに格納されている「試験の点数」を，０点から１０点刻みで集計し，その結果をヒストグラムとして出力する．
    - C2Roomでは，結果表示のためにpyplot.show()は使うことができません．
    - 代わりに， ヒストグラム作成後に，pyplot.savefig("")を実行することで，「画像出力」タブに，ヒストグラムを結果として表示することができます．

```python
import numpy as np
import random
import matplotlib.pyplot as plt

scores = []
for i in range(200):
	scores.append(random.randint(0, 100))
plt.hist(scores, bins=10, rwidth=0.8)
plt.xticks([0, 20, 40, 60, 80, 100])
plt.savefig("")
```

標準ライブラリに含まれるstatisticsモジュールが提供する変数や関数などを用いて，課題「７ー１」で作成したプログラムに次のような機能を追加しなさい．

出力となるヒストグラムに，その作成に用いたデータ（ランダムに生成された「試験の点数」）の平均（算術平均）と分散（標本標準分散）を合わせて表示する．

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import statistics as stats

scores = []
for i in range(200):
	scores.append(random.randint(0, 100))
plt.hist(scores, bins=10, rwidth=0.8)
plt.xticks([0, 20, 40, 60, 80, 100])

a = stats.mean(scores)
b = stats.variance(scores)
plt.text(90, 23, 'average:{}\nvariance:{}'.format(a, b))
plt.savefig("")
```
