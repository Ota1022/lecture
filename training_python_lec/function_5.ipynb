「問５－１」で作成したプログラムを次のように書き換えなさい．  

「問４－２」で示されている形式で標準入力から一連の整数値を（試験の点数として）受け取り，それら整数値を要素とするリストを戻り値として返す関数get_scoresを作成し，それを呼び出すことで「問５－１」と同じ処理を実現する．  
標準入力から与えられる整数値は以下のとおりとする．

「課題５－１」で作成したプログラムを次のように書き換えなさい．

- 関数 calc_daycount_and_dayofweekについて

    - ８桁の整数値を引数とする．引数の上位4桁が「西暦年」を，その次の２桁が「月」を，最下位2桁が「日」を表すものとする．
    - 戻り値は次の要素を返す
        - 0年１月 １日から「引数で与えられた年月日」までの日数（整数値）
    - 引数で与えられる年月日の範囲は，1981年1月1日から2080年12月31日までとする．
    - 1981年から2080年までうるう年はないものとして，引数で与えられる年月日を算出する．（したがって，戻り値は，実際の日数や曜日と異なることになるが，ここではとりあえず構わない．）

- プログラム全体について

    - 入力は，次の２行とする．
        - 第１行　1981年1月１日から2080年12月31日までの任意の年月日
        - 第２行　本日の年月日
    - 出力は，入力で与えられた２つの年月日それぞれの間の日数（差）とする．

```python
a = input()
b = input()

def calc_daycount_and_dayofweek(n):
	month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
	days = 0
	int_year = int(n[:4])
	int_month = int(n[4:6]) - 1
	int_day = int(n[6:8])
	days += 365 * (int_year - 1)
	days += sum(month[:int_month])
	days += int_day
	return days

a_days = calc_daycount_and_dayofweek(a)
b_days = calc_daycount_and_dayofweek(b)
print(b_days - a_days)
```

「課題５－２」で作成したプログラムを，「うるう年」と「1981年1月1日の曜日」を考慮して，正しい曜日や日数差が出力されるよう書き換えなさい．

- 1981年1月1日から2080年12月31日までに限定すると，西暦年号が4で割り切れる年はうるう年である．
- 1981年1月1日は木曜日である．

入力例  

19810101  
20210510  

出力例

Target Thursday  
Today  Monday  
Delta is 14739 day(s)

```python
a = input()
b = input()

def calc_daycount(n):
	month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
	days = 0
	int_year = int(n[:4])
	int_month = int(n[4:6]) - 1
	int_day = int(n[6:8])
	days += 365 * (int_year - 1981)
	days += sum(month[:int_month])
	days += int_day
	if int(n[0:4])%4 == 0 and int(n[4:8]) >= 230:
		uru = int_year // 4 + 1
	else:
		uru = int_year // 4
	days += uru
	return days

def clac_dayofweek(n):
	days = calc_daycount(n)
	dow = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
	remainder = days % 7 -1
	return dow[remainder]

a_dow = clac_dayofweek(a)
b_dow = clac_dayofweek(b)	

a_days = calc_daycount(a)
b_days = calc_daycount(b)

print("Target {}".format(a_dow))
print("Today {}".format(b_dow))
print("Delta is {} day(s)".format(b_days - a_days))
```
